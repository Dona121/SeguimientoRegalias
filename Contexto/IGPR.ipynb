```python
import polars as pl
import polars.selectors as cs
```

```python
ruta_proyectos_regalias = "MatrizSeguimientoEvaluacion.xlsx"

departamento = (
    pl.read_excel(ruta_proyectos_regalias,table_name="MatrizSeguimientoEvaluacion",infer_schema_length=0)
    .select("BPIN","ENTIDAD O SECRETARIA")
    .rename({"ENTIDAD O SECRETARIA":"ENTIDAD"})
)

descentralizadas = (
    pl.read_excel(ruta_proyectos_regalias,table_name="OtrosEjecutoresDescentralizadas",infer_schema_length=0)
    .select("BPIN","EJECUTOR")
    .rename({"EJECUTOR":"ENTIDAD"})
)

municipios = (
    pl.read_excel(ruta_proyectos_regalias,table_name="OtrosEjecutoresMunicipios",infer_schema_length=0)
    .select("BPIN","EJECUTOR")
    .rename({"EJECUTOR":"ENTIDAD"})
)

proyectos_regalias = pl.concat([departamento,descentralizadas,municipios])

# La logica para hacer el join con MatrizSeguimientoEvaluacion.xlsx es a traves del codigo BPIN
```

### Primer trimestre 2025

```python
ruta_primer_trimestre_2025 = "IGPR/Resultados IGPR - I trimestre 2025 06022026.xlsx"
primer_trimestre_2025 = pl.read_excel(ruta_primer_trimestre_2025,table_name="IGPR_I_Trimestre_2025")
columnas_primer_trimestre_2025 = ["BPIN","NOMBRE DEL PROYECTO","IGPR FINAL PROYECTO I TRIM 2025"]

def limpiar_encabezados(trimestre_evaluado) -> pl.Expr:
    return (
        pl.Series(trimestre_evaluado.columns)
        .str.replace_all("_x000a_"," ")
        .str.replace_all("  "," ")
        .str.replace_all(r"\((\d{2}\/\d{2}\/\d{4})\)|(\d{2}\/\d{2}\/\d{4})","")
        .str.replace_all(r"(\*)|(\s\*)","")
        .str.strip_chars()
        .to_list()
    )

limpieza_columnas = limpiar_encabezados(primer_trimestre_2025)


primer_trimestre_2025 = (
    primer_trimestre_2025.rename(dict(zip(primer_trimestre_2025.columns,limpieza_columnas)))
    .filter(pl.col("ENTIDAD EJECUTORA O BENEFICIARIA (A MEDIR)")=="DEPARTAMENTO DE SUCRE")
    .select(columnas_primer_trimestre_2025)
    .with_columns(
        pl.col("BPIN").cast(pl.String),
        pl.lit("PRIMER TRIMESTRE").alias("TRIMESTRE EVALUADO"),
        pl.lit(2025).alias("VIGENCIA"),
    )
    .join(proyectos_regalias,on="BPIN",how="left")
    .rename({"IGPR FINAL PROYECTO I TRIM 2025":"PUNTAJE"})
)
```

### Segundo trimestre 2025

```python
ruta_segundo_trimestre_2025 = "IGPR/Resultados IGPR- II trimestre 2025 06022026.xlsx"
segundo_trimestre_2025 = pl.read_excel(ruta_segundo_trimestre_2025,table_name="IGPR_II_Trimestre_2025")
columnas_segundo_trimestre_2025 = ["BPIN","NOMBRE DEL PROYECTO","IGPR FINAL PROYECTO II TRIM 2025"]

limpieza_columnas = limpiar_encabezados(segundo_trimestre_2025)

segundo_trimestre_2025 = (
    segundo_trimestre_2025.rename(dict(zip(segundo_trimestre_2025.columns,limpieza_columnas)))
    .filter(pl.col("ENTIDAD EJECUTORA O BENEFICIARIA (A MEDIR)")=="DEPARTAMENTO DE SUCRE")
    .select(columnas_segundo_trimestre_2025)
    .with_columns(
        pl.col("BPIN").cast(pl.String),
        pl.lit("SEGUNDO TRIMESTRE").alias("TRIMESTRE EVALUADO"),
        pl.lit(2025).alias("VIGENCIA"),
    )
    .join(proyectos_regalias,on="BPIN",how="left")
    .rename({"IGPR FINAL PROYECTO II TRIM 2025":"PUNTAJE"})
)

# with pl.Config(tbl_rows=50):
#     print(primer_trimestre_2025)
```

### Tercer trimestre 2025

```python
ruta_tercer_trimestre_2025 = "IGPR/Resultados IGPR- III trimestre 2025 11052026.xlsx"
tercer_trimestre_2025 = pl.read_excel(ruta_tercer_trimestre_2025,table_name="IGPR_III_Trimestre_2025")
columnas_tercer_trimestre_2025 = ["BPIN","NOMBRE DEL PROYECTO","IGPR FINAL PROYECTO III TRIM 2025"]

limpieza_columnas = limpiar_encabezados(tercer_trimestre_2025)

tercer_trimestre_2025 = (
    tercer_trimestre_2025.rename(dict(zip(tercer_trimestre_2025.columns,limpieza_columnas)))
    .filter(pl.col("ENTIDAD EJECUTORA O BENEFICIARIA (A MEDIR)")=="DEPARTAMENTO DE SUCRE")
    .select(columnas_tercer_trimestre_2025)
    .with_columns(
        pl.col("BPIN").cast(pl.String),
        pl.lit("TERCER TRIMESTRE").alias("TRIMESTRE EVALUADO"),
        pl.lit(2025).alias("VIGENCIA"),
    )
    .join(proyectos_regalias,on="BPIN",how="left")
    .rename({"IGPR FINAL PROYECTO III TRIM 2025":"PUNTAJE"})
)
```

### Cuatro trimestre 2025

```python
ruta_cuarto_trimestre_2025 = "IGPR/Resultados IGPR - IV trimestre 2025 11052026.xlsx"
cuarto_trimestre_2025 = pl.read_excel(ruta_cuarto_trimestre_2025,table_name="IGPR_IV_Trimestre_2025")
columnas_cuarto_trimestre_2025 = ["BPIN","NOMBRE DEL PROYECTO","IGPR PROYECTO FINAL IV TRIM 2025"]

limpieza_columnas = limpiar_encabezados(cuarto_trimestre_2025)

cuarto_trimestre_2025 = (
    cuarto_trimestre_2025.rename(dict(zip(cuarto_trimestre_2025.columns,limpieza_columnas)))
    .filter(pl.col("ENTIDAD EJECUTORA O BENEFICIARIA (A MEDIR)")=="DEPARTAMENTO DE SUCRE")
    .select(columnas_cuarto_trimestre_2025)
    .with_columns(
        pl.col("BPIN").cast(pl.String),
        pl.lit("CUARTO TRIMESTRE").alias("TRIMESTRE EVALUADO"),
        pl.lit(2025).alias("VIGENCIA"),
    )
    .join(proyectos_regalias,on="BPIN",how="left")
    .rename({"IGPR PROYECTO FINAL IV TRIM 2025":"PUNTAJE"})
)
```

### Primer trimestre 2026

```python
ruta_primer_trimestre_2026 = "IGPR/Resultados IGPR - I trimestre 2026 15052026.xlsx"
primer_trimestre_2026 = pl.read_excel(ruta_primer_trimestre_2026,table_name="IGPR_I_Trimestre_2026")
columnas_primer_trimestre_2026 = ["BPIN","NOMBRE DEL PROYECTO","CLASIFICACIÓN PLAZO PARA EJECUCIÓN",
                                  "INCONSISTENCIAS EN VISITAS","PROYECTOS CON PRESUNTAS IRREGULARIDADES GRAVES",
                                  "MODIFICACIÓN REPORTE EJECUCIÓN","IGPR PROYECTO FINAL"]
# la columna "CLASIFICACIÓN PLAZO PARA EJECUCIÓN" solo se tiene en cuenta para la medición del puntaje de 2026. la puedes
# mostrar solo en los detalles del filtro de este trimestre

limpieza_columnas = limpiar_encabezados(primer_trimestre_2026)

primer_trimestre_2026 = (
    primer_trimestre_2026.rename(dict(zip(primer_trimestre_2026.columns,limpieza_columnas)))
    .filter(pl.col("ENTIDAD EJECUTORA O BENEFICIARIA (A MEDIR)")=="DEPARTAMENTO DE SUCRE")
    .select(columnas_primer_trimestre_2026)
    .with_columns(
        pl.col("BPIN").cast(pl.String),
        pl.col("INCONSISTENCIAS EN VISITAS").str.replace_many({"NO":"0","SI":"1"}).cast(pl.Int32).alias("a"),
        pl.col("PROYECTOS CON PRESUNTAS IRREGULARIDADES GRAVES").str.replace_many({"NO":"0","SI":"1"}).cast(pl.Int32).alias("b"),
        pl.col("MODIFICACIÓN REPORTE EJECUCIÓN").str.replace_many({"NO":"0","SI":"1"}).cast(pl.Int32).alias("c"),
        pl.lit("PRIMER TRIMESTRE").alias("TRIMESTRE EVALUADO"),
        pl.lit(2026).alias("VIGENCIA"),
    ).with_columns(pl.max_horizontal("a","b","c").alias("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE"))
    .join(proyectos_regalias,on="BPIN",how="left")
    .rename({"IGPR PROYECTO FINAL":"PUNTAJE"})
)
```

```python
puntaje = pl.concat(
    [primer_trimestre_2025,segundo_trimestre_2025,tercer_trimestre_2025,cuarto_trimestre_2025],
    how="diagonal"
)

criterios_2026_puntaje_por_vigencia_trimestre = (
    primer_trimestre_2026.group_by("TRIMESTRE EVALUADO","VIGENCIA")
    .agg(pl.col("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE").max(),pl.col("PUNTAJE").mean())
    .with_columns(pl.when(pl.col("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE")==1).then(pl.col("PUNTAJE")-10).otherwise(pl.col("PUNTAJE")))
    .drop("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE")
)

puntaje_por_vigencia_trimestre = (
    pl.concat([puntaje.group_by("TRIMESTRE EVALUADO","VIGENCIA").agg(pl.col("PUNTAJE").mean()),criterios_2026_puntaje_por_vigencia_trimestre])
)

criterios_2026_puntaje_por_entidad_vigencia_trimestre = (
    primer_trimestre_2026.group_by("ENTIDAD","TRIMESTRE EVALUADO","VIGENCIA")
    .agg(pl.col("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE").max(),pl.col("PUNTAJE").mean())
    .with_columns(pl.when(pl.col("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE")==1).then(pl.col("PUNTAJE")-10).otherwise(pl.col("PUNTAJE")))
    .drop("PRESENTA ALGUNA SITUACION QUE AFECTA PUNTAJE")
)

puntaje_por_entidad_vigencia_trimestre = (
    pl.concat([puntaje.group_by("ENTIDAD","TRIMESTRE EVALUADO","VIGENCIA").agg(pl.col("PUNTAJE").mean()),criterios_2026_puntaje_por_entidad_vigencia_trimestre])
)
```
